# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,10])
y = np.array([0,200])
plt.plot(x , y)
plt.show()
```

```python
# plotting without lines

x = np.array([0,10])
y = np.array([0,200])
plt.plot(x , y,"o")
plt.show()
```

```python
# multiple points

x = np.array([0,10,20,25,30,45])
y = np.array([0,25,65,45,78,14])
plt.plot(x , y)
plt.show()
```

```python
# Default x point

y = np.array([0,200])
plt.plot(y)
plt.show()
```

```python
# Markers

x = np.array([0,10,20,25,30,45])
y = np.array([0,25,65,45,78,14])
plt.plot(x , y, marker = "H") # H is a hexagon(symbol)
plt.show()
```

```python
# format string

x = np.array([0,10,20,25,30,45])
y = np.array([0,25,65,45,78,14])
plt.plot(x , y,'o-.g', ms = 15 , mec = "r" , mfc = "y")   # where o is marker , -. is line and m is color of the line and marker
                                                          # ms is the marker size and mec is marker edge color
                                                          # mfc is the marker face color
plt.show()
```

```python
# Line style

x = np.array([0,10,20,25,30,45])
y = np.array([0,25,65,45,78,14])
plt.plot(x , y , linestyle = "dashed" , color = "g" , linewidth = '12')

# labels
plt.title("Data of Example")
plt.xlabel("Example1")
plt.ylabel("Example2" , color = "r")

plt.show()
```

```python
# Grid show

x = np.array([0,10,20,25,30,45])
y = np.array([0,25,65,45,78,14])
plt.plot(x , y)

plt.grid() # axis = "y" , axis = "x" , color , linewidth , linestyle)

plt.show()
```

```python
# Subplot

#Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
# Scatter plot

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)  # color)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)  # color)

plt.show()
```

```python
# Color Map

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
# Bar plot

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color = "r")
plt.show()
```

```python
# pie plot

y = np.array([3, 8, 1, 10])
ylabel = ["A","B","C","D"]
yexplode = [0.2,0,0,0]
plt.pie(y , labels = ylabel , explode = yexplode , shadow = True)
plt.legend(title = "Example")
plt.show()
```
