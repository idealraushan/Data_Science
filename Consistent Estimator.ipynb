# 1. Consistent Estimator

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean = 4.3
sigma = 2
n = 10000
samples = np.random.normal(mean , sigma, (n,n))
estimator = np.mean(samples , axis = 1)
```

```python
#Mean of the estimator
mean_estimator = np.mean(estimator)
```

```python
#Variance of the estimator
variance_estimator = np.var(estimator)
```

```python
#Print Mean and Variance of the estimator
print("Mean of the estimator: ", mean_estimator)
print("variance of the estimator: ", variance_estimator)
```

```python
#Print Mean and Variance of original Distribution
print("Mean of the original distribution: ",mean)
print("variance of the original distribution: ",sigma**2)
```

```python
sns.displot(estimator, bins = 20, kde=True)
plt.title("Histogram of the estimator")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```
